```python
#pip install -U scikit-learn
#Import pandas library for data manipulation and analytics
#Read the csv file and store it in a variable
import pandas as pd
data=pd.read_csv("50_Startups.csv")
data
```

```python
# Using get_dummies, Convert categorical data to meaningful numerical data on data set
# Using drop_first parameter, remove duplicate data from the date set
dataSet=pd.get_dummies(data,drop_first=True)
dataSet
```

```python
#Printout all the columns on the data set
dataSet.columns
```

```python
#Split input data
Independent=dataSet[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
Independent
```

```python
#Split output data
Dependent=dataSet[['Profit']]
Dependent
```

```python
#Split train and test set data using train_test_split method
# the test data should be 1/3 from the data set
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(Independent,Dependent, test_size=1/3,random_state=0)
```

```python
x_train
```

```python
#Create a model using fit method and import require library 'DecisionTreeRegressor'
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='squared_error',splitter='random')
regressor=regressor.fit(x_train,y_train)
```

```python
#Create a graph visualiation based on the data set.
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
# Findout predicted output using test set input data to do evaulation metrics
y_predict=regressor.predict(x_test)
```

```python
y_predict
```

```python
#Evaluate metrics using actual out and predicted out data
from sklearn.metrics import r2_score
result=r2_score(y_test,y_predict)
result
```

```python
# Save the best model and import pickle library
import pickle
file="Finalized_Model_DecisionTree.sav"
pickle.dump(regressor,open(file,'wb'))
```
